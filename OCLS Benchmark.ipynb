```python
import scipy.interpolate as interpolate
import scipy.optimize as optimize

def find_intersetction(x1,y1,x2,y2):
    x1 = np.array(x1)
    y1 = np.array(y1)
    x2 = np.array(x2)
    y2 = np.array(y2)
    
    p1=interpolate.PiecewisePolynomial(x1,y1[:,np.newaxis])
    p2=interpolate.PiecewisePolynomial(x2,y2[:,np.newaxis])
    
    def pdiff(x):
        return p1(x)-p2(x)
    
    xs=np.r_[x1,x2]
    xs.sort()
    x_min=xs.min()
    x_max=xs.max()
    x_mid=xs[:-1]+np.diff(xs)/2
    roots=set()
    for val in x_mid:
        root,infodict,ier,mesg = optimize.fsolve(pdiff,val,full_output=True)
        # ier==1 indicates a root has been found
        if ier==1 and x_min<root<x_max:
            roots.add(root[0])
    roots=list(roots)        
    results = np.row_stack((roots,p1(roots),p2(roots)))
    return results
```

```python
def python_native_solver(T=1.0, Nt=0, CLF=0.95, Nx=100, Omega=[0.0,1.0], I=None, vectorized=True):
    
    c = 1
    
    def flux_l(Q, i,dt,dx):
        return 0.5*(c*Q[i-1]+c*Q[i]) - (dx/(2.0*dt))*(Q[i]-Q[i-1])
    
    def flux_r(Q, i,dt,dx):
        return 0.5*(c*Q[i]+c*Q[i+1]) - (dx/(2.0*dt))*(Q[i+1]-Q[i])
    
    def euler1D(Q, i,dt,dx):
        return Q[i] - (dt/dx)*(flux_r(Q,i,dt,dx)-flux_l(Q,i,dt,dx))
    
    def flux_v(Q,dt,dx):
        return 0.5*(c*Q[:-1]+c*Q[1:]) - (dx/(2.0*dt))*(Q[1:]-Q[:-1])
    
    def euler1D_v(Q,dt,dx):
        F = flux_v(Q,dt,dx)
        return Q[1:-1] - (dt/dx)*(F[1:]-F[:-1])
    
    xv = np.linspace(Omega[0], Omega[1], Nx)
    dx = xv[1]-xv[0]
    
    dt = CLF*dx
    
    Q_0 = np.zeros(Nx+2)
    Q_1 = np.zeros(Nx+2)
    
    Q_0[1:-1] = [I(x) for x in xv]
    if Nt is 0:
        nt = int(ceil(T / dt))+1
    else:
        nt = Nt
    
    start = time.clock()
    for n in range(nt):
        Q_0[:1] = Q_0[-2:-1]
        Q_0[-1:] = Q_0[1:2]
        
        if vectorized is False:
            for i in range(1,Nx+1):
                Q_1[i] = euler1D(Q_0, i, dt, dx)
        else:
            Q_1[1:-1] = euler1D_v(Q_0,dt,dx)
        
        Q_0[:] = Q_1[:]
        
    return Q_0[1:-1], xv, (time.clock()-start)
```

```python
def python_native_solver_2D(T=1.0, Nt=0, CLF=0.95, Nx=100, Ny=100, Omega_x=[0.0,1.0], Omega_y=[0.0,1.0], I=None, vectorized=True):
    
    c = 1
    
    def lax1D(Q, i, j,dt,dx,dy):
        return 0.25*(Q[i+1,j] + Q[i-1,j] + Q[i,j+1] + Q[i,j-1]) - \
                    0.5*(dt/(2.0*dx))*(c*Q[i+1,j] - c*Q[i-1,j]) - \
                    0.5*(dt/(2.0*dy))*(c*Q[i,j+1] - c*Q[i,j-1])
                         
    def lax1D_v(Q, dt,dx,dy):
        return 0.25*(Q[2:,1:-1] + Q[:-2,1:-1] + Q[1:-1,2:] + Q[1:-1,:-2]) - \
                    0.5*(dt/(2.0*dx))*(c*Q[2:,1:-1] - c*Q[:-2,1:-1]) - \
                    0.5*(dt/(2.0*dy))*(c*Q[1:-1,2:] - c*Q[1:-1,:-2])
        
    xv = np.linspace(Omega_x[0], Omega_x[1], Nx)
    yv = np.linspace(Omega_y[0], Omega_y[1], Ny)
    dx = xv[1]-xv[0]
    dy = yv[1]-yv[0]
    
    dt = CLF*max(dx,dy)
    
    Q_0 = np.zeros((Nx+2,Ny+2))
    Q_1 = np.zeros((Nx+2,Ny+2))
    
    Q_0[1:-1,1:-1] = [[I(x,y) for x in xv] for y in yv]
    
    if Nt is 0:
        nt = int(ceil(T / dt))
    else:
        nt = Nt
    
    start = time.clock()
    for n in range(nt):
        Q_0[:1,:]  = Q_0[-2:-1,:]
        Q_0[-1:,:] = Q_0[1:2,:]
        Q_0[:,:1]  = Q_0[:,-2:-1]
        Q_0[:,-1:] = Q_0[:,1:2]

        if vectorized is False:
            for i in range(1,Nx+1):
                for j in range(1,Ny+1):
                    Q_1[i,j] = lax1D(Q_0, i, j, dt, dx, dy)
        else:
            Q_1[1:-1,1:-1] = lax1D_v(Q_0,dt,dx,dy)
        
        Q_0[:,:] = Q_1[:,:]
        
    return Q_0[1:-1,1:-1], xv, yv, (time.clock()-start)
```

```python
def pyopencl_solver(T=1.0, Nt=0, CLF=0.95, Nx=100, Omega=[0.0,1.0], I=None):
    
    ctx = cl.create_some_context()
    queue = cl.CommandQueue(ctx)
    
    xv = np.linspace(Omega[0], Omega[1], Nx)
    
    dx = xv[1]-xv[0]
    dt = CLF*dx
    
    Q_ = np.zeros(Nx+2).astype(np.float32)
    Q_[1:-1] = [I(x) for x in xv]
    
    mf = cl.mem_flags
    Q_0 = cl.Buffer(ctx, mf.READ_WRITE | mf.COPY_HOST_PTR, hostbuf=Q_)
    Q_1 = cl.Buffer(ctx, mf.READ_WRITE, Q_.nbytes)
    F = cl.Buffer(ctx, mf.READ_WRITE, Q_.nbytes)
    
    prg = cl.Program(ctx, """
    __constant float c = 1.0;
    
    __kernel void bounds(__global float *Q, int const Nx){
        Q[0] = Q[Nx];
        Q[Nx+1] = Q[1];
    }
    
    __kernel void num_flux(__global float *Q_0, __global float *F, float dt, float dx) {      
        int gid = get_global_id(0);
        
        float q_r = Q_0[gid+1];
        float q_l = Q_0[gid];
      
        F[gid] = 0.5f*(c*q_l+c*q_r) - (dx/(2.0f*dt))*(q_r-q_l);
    }
    
    __kernel void euler(__global float *Q_0, __global float *F, __global float *Q_1, float dt, float dx){
        int gid = get_global_id(0)+1;
        
        float f_r = F[gid];
        float f_l = F[gid-1]
        float q = Q_0[gid];
        
        Q_1[gid] = q - (dt/dx)*(f_r-f_l);
    }
    """).build()
    
    if Nt is 0:
        nt = int(ceil(T / dt))
    else:
        nt = Nt
    
    start = time.clock()
    for n in range(nt):
        prg.bounds(queue, 1, None, Q_0, Nx)
        prg.num_flux(queue, Nx+1, None, Q_0, F, dt, dx)
        prg.euler(queue, Nx, None, Q_0, F, Q_1, dt, dx)
        
        cl.enqueue_copy(queue, Q_0, Q_1)
        
    cl.enqueue_copy(queue, Q_, Q_0)
    return Q_[1:-1], xv, (time.clock()-start)
```

```python
import ocls_framework as ocls
import time

def ocls_solver(T=1.0, Nt=0, CLF=0.95, Nx=100, Omega=[0.0,1.0], I=None, device="GPU"):
    
    framework = None
    if device == "GPU":
        framework = ocls.Framework(ocls.Type.GPU)
    else:
        framework = ocls.Framework(ocls.Type.CPU)
    
    d = ocls.Domain(1)
    d.setMesh(x=Nx)
    d.setBounds(x=Omega)
    d.setGhostCells(1)
    
    src_boundary = framework.loadFromString("""
    void periodic(float* Q){
        if(LEFT)
            store(Q, fetch_offset(Q, cell.x, Nx), cell.x);
        else
            store(Q, fetch_offset(Q, cell.x, -Nx), cell.x);  
    }""")
    src_flux = framework.loadFromString("""
    float flux(float q){
        const float c = 1.0f;
        return c*q;
    }

    float lax_friedrich(float* Q, float dt, float dx){
        float q = fetch(Q, cell.x);
        float q_p = fetch_offset(Q, cell.x, 1);

        return 0.5f*(flux(q)+flux(q_p)) - (dx/(2.0f*dt))*(q_p-q);
    }
    """)
    
    src_inter = framework.loadFromString("""
    float forward_euler(float* Q, float* flux, float dt, float dx){
        float q = fetch(Q, cell.x);
        float F_m = fetch_offset(flux, cell.x,-1);
        float F_p = fetch(flux, cell.x);
    
        return q - (dt/dx)*(F_p-F_m);
    }
    """)
    
    q = framework.createData(d, "q")
    
    if I is not None:
        q.set(I)
    
    on_bounds = framework.createBoundaryFunction(d, src_boundary, "periodic")
    time_integrator = framework.createIntegratorFunction(d, src_inter, "forward_euler")
    flux_function = framework.createFluxFunction(d, src_flux, "lax_friedrich")
    
    dx = d.dx()
    dt = CLF*dx
    
    if Nt is 0:
        nt = int(ceil(T / dt))
    else:
        nt = Nt
    
    start = time.clock()
    for i in range(nt):
        on_bounds(q)
        flux = flux_function(q, dt, dx)
        q_1 = time_integrator(q, flux, dt, dx)
        
        q.copy(q_1)
            
    return q.toArray()[1:-1].copy(), d.toLinspace(), (time.clock()-start)

def I_(x):
    if x[0] > 0.5  and x[0] < 0.7:
        return 1.0
    else:
        return 0.0
```

```python
import ocls_framework as ocls
import time

def ocls_solver_2D(T=1.0, Nt=0, CLF=0.95, Nx=100, Ny=100, Omega_x=[0.0,1.0], Omega_y=[0.0,1.0], I=None, device="GPU"):
    
    framework = None
    if device == "GPU":
        framework = ocls.Framework(ocls.Type.GPU)
    else:
        framework = ocls.Framework(ocls.Type.CPU)
    
    d = ocls.Domain(2)
    d.setMesh(x=Nx,y=Ny)
    d.setBounds(x=Omega_x, y=Omega_y)
    d.setGhostCells(1)
    
    src_boundary = framework.loadFromString("""
    void periodic(float* Q){
        if(LEFT)
            store(Q, fetch_offset(Q, cell.xy, (int2)(-Nx,0)), cell.xy);
        else if(RIGHT)
            store(Q, fetch_offset(Q, cell.xy, (int2)(Nx,0)), cell.xy);  
        else if(NORTH)
            store(Q, fetch_offset(Q, cell.xy, (int2)(0,Ny)), cell.xy);
        else if(SOUTH)
            store(Q, fetch_offset(Q, cell.xy, (int2)(0,-Ny)), cell.xy);   
    }
    """)
    
    src_inter = framework.loadFromString("""
    float flux(float q){
        const float c = 1.0f;
        return c*q;
    }
    
    float lax_friedrichs(float* Q, float dt, float dx, float dy){
        float q_l = fetch_offset(Q, cell.xy,(int2)(-1, 0));
        float q_r = fetch_offset(Q, cell.xy,(int2)( 1, 0));
        float q_n = fetch_offset(Q, cell.xy,(int2)( 0,-1));
        float q_s = fetch_offset(Q, cell.xy,(int2)( 0, 1));
    
        return 0.25f*(q_r+q_l+q_n+q_s)-0.5f*(dt/(2.0f*dx))*(flux(q_l)-flux(q_r))-0.5f*(dt/(2.0f*dy))*(flux(q_n)-flux(q_s));
    }
    """)
    
    q = framework.createData(d, "q")
    
    if I is not None:
        q.set(I)
    
    on_bounds = framework.createBoundaryFunction(d, src_boundary, "periodic")
    time_integrator = framework.createIntegratorFunction(d, src_inter, "lax_friedrichs")
    
    dx = d.dx()
    dy = d.dy()
    dt = CLF*max(dx,dy)
    
    if Nt is 0:
        nt = int(ceil(T / dt))
    else:
        nt = Nt
    
    start = time.clock()
    for i in range(nt):
        on_bounds(q)
        q_1 = time_integrator(q, dt, dx, dy)
        
        q.copy(q_1)
            
    return q.toArray()[1:-1, 1:-1].copy(), d.toLinspace(), (time.clock()-start)

def I_2D_(x,y):
    if x[0] > 0.5  and x[0] < 0.7:
        return 1.0
    else:
        return 0.0
```

```python
def clawpack_solver(T=1.0, Nt=0, CLF=0.95, Nx=100, Omega=[0.0,1.0], I=None):
    from clawpack import pyclaw
    from clawpack import riemann
        
    
    #solver = pyclaw.SharpClawSolver1D(riemann.advection_1D)
    #solver.time_integrator = 'Euler'
    #solver.lim_type = 1
    
    solver = pyclaw.ClawSolver1D(riemann.advection_1D)
    #solver = pyclaw.ClawSolver1D()
    solver.order = 1
    solver.limiters = 0
    
    solver.cfl_max = CLF
    
    solver.bc_lower[0] = pyclaw.BC.periodic
    solver.bc_upper[0] = pyclaw.BC.periodic

    x = pyclaw.Dimension(Omega[0],Omega[1],Nx,name='x')
    domain = pyclaw.Domain(x)
    state = pyclaw.State(domain,1)

    state.problem_data['u'] = 1.  # Advection velocity

    # Initial data
    xc = state.grid.x.centers
    state.q[0,:] = [I(x) for x in xc]

    claw = pyclaw.Controller()
    #claw.keep_copy = True
    claw.solution = pyclaw.Solution(state,domain)
    claw.solver = solver

    claw.tfinal =1.0
    claw.verbosity = 0

    start = time.clock()
    claw.run()
    
    return state.q[0], xc, (time.clock()-start)
```

```python
def clawpack_solver_2D(T=1.0, Nt=0, CLF=0.95, Nx=100, Ny=100, Omega_x=[0.0,1.0], Omega_y=[0.0,1.0], I=None):
    from clawpack import pyclaw
    from clawpack import riemann
        
    
    #solver = pyclaw.SharpClawSolver1D(riemann.advection_1D)
    #solver.time_integrator = 'Euler'
    #solver.lim_type = 1
    
    solver = pyclaw.ClawSolver2D(riemann.advection_2D)
    #solver = pyclaw.ClawSolver2D()
    solver.order = 1
    solver.limiters = 0
    
    solver.cfl_max = CLF
    
    solver.bc_lower[0] = pyclaw.BC.periodic
    solver.bc_lower[1] = pyclaw.BC.periodic
    solver.bc_upper[0] = pyclaw.BC.periodic
    solver.bc_upper[1] = pyclaw.BC.periodic

    x = pyclaw.Dimension(Omega_x[0],Omega_x[1],Nx,name='x')
    y = pyclaw.Dimension(Omega_y[0],Omega_y[1],Ny,name='y')
    domain = pyclaw.Domain((x,y))
    state = pyclaw.State(domain,1)

    state.problem_data['u'] = 1.  # Advection velocity
    state.problem_data['v'] = 1.

    # Initial data
    xc = state.grid.x.centers
    yc = state.grid.y.centers
    state.q[0,:,:] = [[I(x,y) for x in xc] for y in yc]

    claw = pyclaw.Controller()
    #claw.keep_copy = True
    claw.solution = pyclaw.Solution(state,domain)
    claw.solver = solver

    claw.tfinal =1.0
    claw.verbosity = 0

    start = time.clock()
    claw.run()
    
    return state.q[0], xc, yc, (time.clock()-start)
```
