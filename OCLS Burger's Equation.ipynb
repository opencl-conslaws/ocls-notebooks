# Burger's Equation

Inviscid Burger's Equation in 1D written out in conservative form:

\begin{aligned}
    \frac{\partial q}{\partial t} + \frac{1}{2}\frac{\partial}{\partial x}(q^2) = 0
\end{aligned}

Python string with device code

```python
boundary_cons = """
void periodic(float* Q){
    if(LEFT)
        store(Q, fetch_offset(Q, cell.x, Nx), cell.x);
    else
        store(Q, fetch_offset(Q, cell.x, -Nx), cell.x);  
}

void absorbing(float* Q){
    store(Q, fetch_mirror(Q, cell.x), cell.x);
}
"""
```

```python
integrators = """
float forward_euler(float* Q, float* flux, float dt, float dx){
    float q = fetch(Q, cell.x);
    float F_m = fetch_offset(flux, cell.x,-1);
    float F_p = fetch(flux, cell.x);
    
    return q - (dt/dx)*(F_p-F_m);
}
"""
```

```python
numerical_flux = """
float flux(float q){
    return 0.5f*q*q;
}

float lax_friedrich(float* Q, float dt, float dx){
    float q = fetch(Q, cell.x);
    float q_p = fetch_offset(Q, cell.x, 1);
    
    return 0.5f*(flux(q)+flux(q_p)) - (dx/(2.0f*dt))*(q_p-q);
}

float lax_wendroff(float* Q, float dt, float dx){
    float q = fetch(Q, cell.x);
    float q_p = fetch_offset(Q, cell.x, 1);
    
    return flux(0.5f*(q+q_p) - (0.5f*dt/dx)*(flux(q_p)-flux(q)));
}
"""
```

Solver function for 1D burger

```python
def burger1d_solver(T, d, dt=0, CFL=0.9, action=None,
                    I=None, B="periodic", 
                    flux_function="lax_friedrich",
                    integrator="forward_euler"):
    framework = ocls.Framework()
    visualizer = viz.init(framework)
    
    src_boundary = framework.loadFromString(boundary_cons)
    src_integrators = framework.loadFromString(integrators)
    src_numerics = framework.loadFromString(numerical_flux)
    
    q = framework.createData(d, "q")
    if I is not None:
        q.set(I)
    
    on_bounds = framework.createBoundaryFunction(d, src_boundary, B)
    time_integrator = framework.createIntegratorFunction(d, src_integrators, integrator)
    numerical_flux_func = framework.createFluxFunction(d, src_numerics, flux_function)
    
    dx = d.dx()
    if dt is 0:
        dt = CFL*dx/q.max()
    
    x = d.toLinspace()
    t = 0.0
    nt = int(ceil(T / dt))
    
    # Step 0
    if action is not None:
        action(x,q,t,0)
    
    start = time.clock()
    
    for i in range(nt):
        on_bounds(q)
        flux = numerical_flux_func(q, dt, dx)
        q_1 = time_integrator(q, flux, dt, dx)
        
        q.copy(q_1)
        t = dt*(i+1)
        
        if action is not None:
            action(x,q,t,i+1)
            
    end = time.clock()
            
    return dt, (end-start)
```

Plot some graphs!

```python
T = 1.0
Nx = 150

Omega = [0.0,1.5]

d = ocls.Domain(1)
d.setMesh(x=Nx)
d.setBounds(x=Omega)
d.setGhostCells(1)

q_I = x_I = None

def store_exact(x, q, t, n):
    global q_I, x_I
    if t > T:
        x_I = x.copy()
        q_I = q.toArray()[1:-1].copy()

def plot(x, q, t, n):
    global q_I, x_I, T, title
    #if n is 0:
    #    q_I = q.toArray()[1:-1].copy()
    if t > T:
        qy = q.toArray()[1:-1]
        axes.plot(x, qy, 'bo')
        axes.plot(x_I, q_I, 'r-')
        axes.set_ylim(-0.2 ,1.2)
        axes.set_xlabel('x')
        axes.set_ylabel('q')
        axes.set_title('{0}'.format(title))

import math
def I(pos):
    if pos[0] > 0.1 and pos[0] < 0.4:
        return math.sin(((pos[0]-0.1)/0.3)*math.pi)**2
    elif pos[0] > 0.6 and pos[0] < 0.9:
        return 1.0
    else:
        return 0.0
    
    
large_d = ocls.Domain(1)
large_d.setMesh(x=10000)
large_d.setBounds(x=Omega)
large_d.setGhostCells(1)
burger1d_solver(T, large_d, dt=0.95*large_d.dx(), I=I, action=store_exact,
                          flux_function="lax_friedrich")
```

```python
fig, axe = plt.subplots(1,2, figsize=(16,6))

title = "Lax-Friedrich"
axes = axe[0]
burger1d_solver(T, d, dt=0.95*d.dx(), I=I, action=plot,
                          flux_function="lax_friedrich")

title = "Lax-Wendroff"
axes = axe[1]
burger1d_solver(T, d, dt=0.95*d.dx(), I=I, action=plot,
                          flux_function="lax_wendroff")
```

## High-Resolution Schemes

```python
numerical_flux = """

float flux(float q){
    return 0.5f*q*q;
}

float linear_slope(float* Q, int i){
    float q = fetch(Q, i);
    float q_p = fetch_offset(Q, i, 1);
    float q_m = fetch_offset(Q, i, -1);
    
    return minmod_slope(q, q_m, q_p);
    //return minmodAB(q-q_m,q_p-q);
}

float central_upwind(float* Q){
    float q_r = fetch_offset(Q, cell.x, 1); //"Right" cell in Riemann problem
    float q_l = fetch_offset(Q, cell.x, 0); //"Left" --"--
    
    //Reconstructed slopes
    float s_r = linear_slope(Q, cell.x+1); 
    float s_l = linear_slope(Q, cell.x);
    
    float q_rr = q_r - 0.5f*s_r;
    float q_ll = q_l + 0.5f*s_l;
    
    float ap, am;
    
    // ????? 
    ap  = 0.5*(q_l+q_r);
    am  = -1.0f; // Seems to work, no idea why... #confuse
    // ?????

    return ((ap*flux(q_ll) - am*flux(q_rr)) + (ap*am)*(q_rr-q_ll))/(ap-am);
}

float MUSCL(float* Q){
    float q_r = fetch_offset(Q, cell.x, 1); //"Right" cell in Riemann problem
    float q_l = fetch_offset(Q, cell.x, 0); //"Left" --"--
    
    //Reconstructed slopes
    float s_r = linear_slope(Q, cell.x+1); 
    float s_l = linear_slope(Q, cell.x);
    
    float q_rr = q_r - 0.5f*s_r;
    float q_ll = q_l + 0.5f*s_l;
    
    float ap;
    ap  = 0.5*(q_l+q_r);
    
    // Local Lax flux
    return 0.5f*((flux(q_ll)+flux(q_rr))-ap*(q_rr-q_ll));
}
"""
```

```python
def burger1d_solver(T, d, dt=0, CFL=0.5, action=None,
                    I=None, B="periodic", 
                    flux_function="central_upwind",
                    integrator="forward_euler"):
    framework = ocls.Framework()
    visualizer = viz.init(framework)
    
    
    src_boundary = framework.loadFromString(boundary_cons)
    src_integrators = framework.loadFromString(integrators)
    src_numerics = framework.loadFromString(numerical_flux)
    
    q = framework.createData(d, "q")
    if I is not None:
        q.set(I)
    
    on_bounds = framework.createBoundaryFunction(d, src_boundary, B)
    time_integrator = framework.createIntegratorFunction(d, src_integrators, integrator)
    numerical_flux_func = framework.createFluxFunction(d, src_numerics, flux_function)
    
    dx = d.dx()
    if dt is 0:
        dt = CFL*dx/q.max()
    
    x = d.toLinspace()
    t = 0.0
    nt = int(ceil(T / dt))
    
    # Step 0
    if action is not None:
        action(x,q,t,0)
    
    start = time.clock()
    
    for i in range(nt):
        on_bounds(q)
        flux = numerical_flux_func(q)
        q_1 = time_integrator(q, flux, dt, dx)
        
        q.copy(q_1)
        t = dt*(i+1)
        
        if action is not None:
            action(x,q,t,i+1)
            
    end = time.clock()
            
    return dt, (end-start)
```
