```python
import ocls_framework as ocls
import ocls_framework.compiler as ocls_comp
from ocls_framework.functions import boundary as ocls_bounds

def solve():
    framework = ocls.Framework()
    
    domain = ocls.Domain(1)
    domain.setMesh(x=100)
    domain.setBounds(x=[0.0,1.0])
    domain.setGhostCells(1)
    
    c = ocls_comp.Compiler(framework, domain, verbose=True)
    
    @c.boundary(Q='data')
    def on_bounds(Q):
        if LEFT:
            Q[cell.x] = Q[cell.x+Nx]
        else:
            Q[cell.x] = Q[cell.x-Nx]
    
    @c.device(q='float')
    def fflux(q):
        v = new_float(1.0*q)
        return v
    
    @c.flux(Q='data', dt='float', dx='float')
    def lax(Q, dt, dx):
        q_l = Q[cell.x]
        q_r = Q[cell.x+1]
    
        F = new_float(0.5*(fflux(q_l)+fflux(q_r)) - (dx/(2.0*dt))*(q_r-q_l))
        return F
    
    @c.integrator(Q='data', f='data', 
                  dt='float', dx='float')
    def euler(Q, f, dt, dx):
        q   = Q[cell.x]
        f_m = f[cell.x-1]
        f_p = f[cell.x]
    
        q0 = new_float(q - (dt/dx)*(f_p-f_m))
        return q0
    
    q = ocls.Data(framework,domain)
    
    def I(x):
        if x[0] > 0.1 and x[0] < 0.4:
            return 1.0
        else:
            return 0.0
    q.set(I)
    
    dx = domain.dx()
    dt = 0.95*dx
    
    T = 1.0
    t = 0.0
    
    while(t < T):
        on_bounds(q)
        f = lax(q, dt, dx)
        q0 = euler(q, f, dt, dx)
    
        q.copy(q0)
    
        t+=dt
    
    return q.toArray()[1:-1].copy(), domain.toLinspace()

q, x = solve()

import matplotlib.pyplot as plt
plt.plot(x,q, 'b-o')
plt.show()
```

```python
import scipy.interpolate as interpolate
import scipy.optimize as optimize
import numpy as np

x1=np.array([1.4,2.1,3,5.9,8,9,23])
y1=np.array([2.3,3.1,1,3.9,8,9,11])
x2=np.array([1,2,3,4,6,8,9])
y2=np.array([4,12,7,1,6.3,8.5,12])    

p1=interpolate.PiecewisePolynomial(x1,y1[:,np.newaxis])
p2=interpolate.PiecewisePolynomial(x2,y2[:,np.newaxis])

def pdiff(x):
    return p1(x)-p2(x)

xs=np.r_[x1,x2]
xs.sort()
x_min=xs.min()
x_max=xs.max()
x_mid=xs[:-1]+np.diff(xs)/2
roots=set()
for val in x_mid:
    root,infodict,ier,mesg = optimize.fsolve(pdiff,val,full_output=True)
    # ier==1 indicates a root has been found
    if ier==1 and x_min<root<x_max:
        roots.add(root[0])
roots=list(roots)        
results = np.row_stack((roots,p1(roots),p2(roots)))
print results

import matplotlib.pyplot as plt

plt.plot(x1,y1)
plt.plot(x2,y2)
plt.plot(results[0],results[1],'ro')

for i in range(0, len(results)-1):
    x = results[0][i]
    y = results[1][i]
    
    plt.annotate('y={0}'.format(y), xy=(x, y), xytext=(x-5, y+5),
            bbox = dict(boxstyle = 'round,pad=0.5', fc = 'yellow', alpha = 0.5),
            arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))

plt.show()
```

```python
import numpy as np
Q = np.array([[1.0,0.5,0.1],
             [1.0,0.5,0.1],
             [1.0,0.5,0.1]])
Q = Q.transpose()

c = 1.0
dx = 0.01
dy = 0.01
dt = 0.45*dx

def F_Flux_L(i,j):
    return 0.5*(c*Q[i-1,j]+c*Q[i,j]) - (dx/2.0*dt)*(Q[i,j]-Q[i-1,j])

def F_Flux_R(i,j):
    return 0.5*(c*Q[i,j]+c*Q[i+1,j]) - (dx/2.0*dt)*(Q[i+1,j]-Q[i,j])

def G_Flux_L(i,j):
    return 0.5*(c*Q[i,j-1]+c*Q[i,j]) - (dy/2.0*dt)*(Q[i,j]-Q[i,j-1])

def G_Flux_R(i,j):
    return 0.5*(c*Q[i,j]+c*Q[i,j+1]) - (dy/2.0*dt)*(Q[i,j+1]-Q[i,j])

def euler2D(i,j):
    L = -((F_Flux_R(i,j)-F_Flux_L(i,j))/dx+(G_Flux_R(i,j)-G_Flux_L(i,j))/dy)
    return Q[i,j] + dt*L;

print "Q_1 eul2D ", euler2D(1,1)


def lax2D(i,j):
    ## R, L, N, S
    return 0.25*(Q[i+1,j]+Q[i-1,j]+Q[i,j-1]+Q[i,j+1]) -\
                0.5*(dt/dx)*(c*Q[i+1,j]-c*Q[i-1,j])-\
                0.5*(dt/dy)*(c*Q[i,j+1]-c*Q[i,j-1])
    
print "Q_1 lax2D", lax2D(1,1)

dt = 0.95*dx

def euler1D(i,j):
    L = -(F_Flux_R(i,j)-F_Flux_L(i,j))/dx
    return Q[i,j] + dt*L;

print "Q_1 eul1D", euler1D(1,1)
    
def lax1D(i,j):
    return 0.5*(Q[i+1,j]+Q[i-1,j]) -\
        (dt/2.0*dx)*(c*Q[i+1,j]-c*Q[i-1,j])
        
print "Q_1 lax1D", lax1D(1,1)
```

```python
def createData(d, func="gaussian1d", use_numpy=True):
    source = """
    float gauss(float x, float b, float c){
        return (((x-b)*(x-b))/2.0f*c*c);
    }
    
    float gaussian1d(void){
        float val = 0.0f;
        val += gauss(pos.x,-0.0f,10.5f);
        return 10.0f*exp(-val)+2.0f;
    }
    float gaussian2d(void){
        float val = 0.0f;
        val += gauss(pos.x,-0.0f,10.5f);
        val += gauss(pos.y,-0.0f,10.5f);
        return 10.0f*exp(-val)+2.0f;
    }
    float gaussian3d(void){
        float val = 0.0f;
        float val2 = 0.0f;
    
        val += gauss(pos.x,-1.0f,2.5f);
        val += gauss(pos.y,-1.0f,2.5f);
        val += gauss(pos.z,-1.0f,2.5f);
    
        val2 += gauss(pos.x,1.0f,2.5f);
        val2 += gauss(pos.y,1.0f,2.5f);
        val2 += gauss(pos.z,1.0f,2.5f);
    
        return exp(-val)+exp(-val2);
    }
    
    float out_constant(void){
        return 1.0f;
    }
    """
    
    framework = None #Cleanup bug workaround
    framework = ocls.Framework()
    src = framework.loadFromString(source)

    q = framework.createData(d, "q", src, func)
    arr = q.toArray().copy()
    return arr
```

```python
import os
def create_movie_html(prefix):
    codec2ext = dict(flv='flv', libx264='mp4', libvpx='webm',
                     libtheora='ogg')
    fps = 24
    filespec = '/tmp/{0}_frame_%04d.png'.format(prefix)
    movie_program = 'avconv'  # or 'ffmpeg'
    for codec in codec2ext:
        ext = codec2ext[codec]
        cmd = '%(movie_program)s -r %(fps)d -i %(filespec)s -y '\
              '-vcodec %(codec)s /tmp/%(prefix)s_movie.%(ext)s' % vars()
        #print cmd
        os.system(cmd)
    
    from base64 import b64encode
    
    video_tag = '<video controls>'
    for codec in codec2ext:
        ext = codec2ext[codec]
        video = open("/tmp/%(prefix)s_movie.%(ext)s" % vars(), "rb").read()
        video_encoded = b64encode(video)
        video_tag += ('<source src="data:video/{1};base64,{0}" type="video/{1}">'
                         .format(video_encoded, ext))
    video_tag += 'No video support</video>'
    return video_tag  
```

```python
burger1d = """
float gauss(float x, float b, float c){
    return (((x-b)*(x-b))/2.0f*c*c);
}

float gaussian(void){
    float val = 0.0f;
    val += gauss(pos.x,-0.8f,10.5f);
    return 10.0f*exp(-val)+2.0f;
}

void on_bounds(float* Q){
    store(Q, fetch_mirror(Q, cell.x), cell.x);
}

float flux(float q){
    return 0.5f*q*q;
}

float lax_friedrichs(float* Q, float dt, float dx){
    float q_m = fetch_offset(Q, cell.x,-1);
    float q_p = fetch_offset(Q, cell.x, 1);
    
    return 0.5f*(q_m+q_p) - (dt/dx)*(flux(q_p)-flux(q_m));
}
"""
```

```python
import ocls_framework as ocls

framework = None #Cleanup bug workaround
framework = ocls.Framework()

d = ocls.Domain(1)
d.setDimensions(1)
d.setMesh(x=200)
d.setBounds(x=[-1.0,1.0])
d.setGhostCells(1)

src = framework.loadFromString(burger1d)

q = framework.createData(d, "q", src, "gaussian")

on_bounds = framework.createBoundaryFunction(d, src, "on_bounds")
lax_friedrichs = framework.createIntegratorFunction(d, src, "lax_friedrichs")

dx = d.dx()
dt = 0.5*dx/q.max()

y_min = q.min()
y_max = q.max()

import matplotlib.pyplot as plt
import os

def plot(x, q, t, n):
    qy = [q[i] for i in range(2,202)]
    plt.figure()
    plt.plot(x, qy, 'b')
    plt.axis([-1.0,1.0, y_min,y_max])
    plt.xlabel('x')
    plt.ylabel('q')
    plt.title('Burger 1D @ t={0}'.format(t))
    #plt.show()
    plt.savefig('/tmp/burger1d_frame_%04d.png' % n)  # for movie making

#For plotting
x = d.toLinspace()

T = 0.2
t = 0.0
plot(x, q, t, 0)

nt = int(ceil(T / dt))
for i in range(nt):
    on_bounds(q)
    q_1 = lax_friedrichs(q, dt, dx)
    
    q.copy(q_1)
    t = dt*(i+1)
    
    plot(x, q, t, i+1)

plt.close('all')
framework = None #Cleanup bug workaround
```
